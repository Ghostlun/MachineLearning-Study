```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Position_Salaries.csv")
X = dataset.iloc[:, 1: -1].values
y = dataset.iloc[:,-1].values
```

```python
#Traning the Decision Tree Regression model on whole dataset
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(X, y)
```

```python
# Prediciting a new result
regressor.predict([[6.5]])
```

```python
# Visualizing the decision tree regeression result (higher resolution)
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = "red")
plt.plot(X_grid, regressor.predict(X_grid), color = "blue")
plt.title("Decision Tree instuition")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

```python
# Visualizing the decision tree regeression result (lower resolution))
plt.plot(X, regressor.predict(X), color = "blue")
plt.title("Decision Tree instuition")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```
